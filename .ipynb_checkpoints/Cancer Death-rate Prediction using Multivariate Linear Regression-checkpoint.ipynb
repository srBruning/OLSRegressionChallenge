### Correlation

```python
# Checking the correlation
```

```python
sns.set(rc={'figure.figsize':(25,25)})
sns.heatmap(cancer.corr(), annot=True, cmap='Reds')
```

```python
# Subsetting the columns by removing highly correlated variables
```

```python
new_feature = ['target_deathrate','avganncount', 'incidencerate',
       'medincome', 'popest2015', 'povertypercent', 'studypercap',
       'medianage', 'medianagefemale',
       'percentmarried', 'pctnohs18_24', 'pcths18_24', 'pctbachdeg18_24',
       'pcths25_over', 'pctbachdeg25_over', 'pctemployed16_over',
       'pctunemployed16_over', 'pctprivatecoverage',
        'pctpubliccoverage', 
       'pctwhite', 'pctblack', 'pctasian', 'pctotherrace',
        'birthrate']

sns.heatmap(cancer[new_feature].corr(), annot=True, cmap='Reds')
```

### Model Building

```python
# Dividing the data into test and train
```

```python
from sklearn.model_selection import train_test_split

train ,test = train_test_split(cancer[new_feature],test_size=0.3,random_state = 123 ) 
```

```python
# Checking the dimensions of test and train
```

```python
train.shape
```

```python
test.shape
```

```python
train.columns
```

```python
import statsmodels.formula.api as smf
```

```python
# Linear regression model on the train dataset
```

```python
lm=smf.ols('target_deathrate~avganncount+incidencerate+medincome+popest2015+povertypercent+studypercap+medianage+medianagefemale+percentmarried+pctnohs18_24+pcths18_24+pctbachdeg18_24+pcths25_over+pctbachdeg25_over+pctemployed16_over+pctunemployed16_over+pctprivatecoverage+pctpubliccoverage+pctwhite+pctblack+pctasian+pctotherrace+birthrate', train).fit()
```

```python
print(lm.summary())
```

```python
# Applying the trained linear regression model on the test data
```

```python
pred = lm.predict(test)
pred.head(5)
```

```python
# Model evaluation 
```

```python
from sklearn import metrics
import numpy as np
print('MAE:', metrics.mean_absolute_error(test.target_deathrate, pred))
print('MSE:', metrics.mean_squared_error(test.target_deathrate, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(test.target_deathrate, pred)))
#print 'R-Squre:', metrics.r2_score(test.ln_sales_in_thousands, pred)
```

```python
# Builiding another model by reducing the number of variables for which p-value is less than 0.05
```

```python
lm1=smf.ols('target_deathrate~incidencerate+medianagefemale+pctnohs18_24+pcths18_24+pcths25_over+pctbachdeg25_over+pctunemployed16_over+pctprivatecoverage+pctwhite+pctotherrace+birthrate', train).fit()
```

```python
print(lm1.summary())
```

```python
# Applying the newly trained model on the test data
```

```python
pred1 = lm1.predict(test)
pred1.head(5)
```

```python
# Again checking the model evaluation
```

```python
print('MAE:', metrics.mean_absolute_error(test.target_deathrate, pred1))
print('MSE:', metrics.mean_squared_error(test.target_deathrate, pred1))
print('RMSE:', np.sqrt(metrics.mean_squared_error(test.target_deathrate, pred1)))
#print 'R-Squre:', metrics.r2_score(test.ln_sales_in_thousands, pred)
```
